```python
# ORIGINAL - GET BEST MODEL RESULTS
results_data_generalization = []
for NET in NETS:
    if NET == 'dilation':
        import experiments.dilation as experiments
    elif NET == 'segnet':
        import experiments.segnet as experiments
    elif NET == 'lstm':
        import experiments.lstm as experiments
    elif NET == 'coloring':
        import experiments.coloring as experiments
    elif NET == 'crossing':
        import experiments.crossing as experiments
    elif NET == 'unet':
        import experiments.unet as experiments
    elif NET == 'multi_lstm':
        import experiments.multi_lstm as experiments
    elif NET == 'multi_lstm_init':
        import experiments.multi_lstm_init as experiments
    elif NET == 'FF':
        import experiments.FF as experiments
    elif NET == 'optimal_lstm':
        import experiments.optimal_lstm as experiments

    output_path = '/om/user/shobhita/src/insideness/' + NET + '/'
    run_opt = experiments.get_experiments(output_path)
    opt_data = datasets.get_datasets(output_path)

    NUM_COMPLEXITIES = 6
    results_data_generalization.append([])
    for opt in run_opt[1:]:
        data_point = {}

        data_point["dataset_complexity"] = opt.dataset.complexity
        data_point["dataset_strict_complexity"] = opt.dataset.complexity_strict
        data_point["im_size"] = opt.dataset.image_size
        data_point["id"] = opt.ID
        
        if not os.path.isfile(opt.log_dir_base + opt.name + '/results/generalization_accuracy_60_full1.pkl'):
            data_point["results"] = "empty"
            print(opt.log_dir_base + opt.name + '/results/generalization_accuracy_60_full1.pkl')
        else:
            with open(opt.log_dir_base + opt.name + '/results/generalization_accuracy_60_full1.pkl', 'rb') as f:
                data_point["results"] = pickle.load(f)

        results_data_generalization[-1].append(copy.deepcopy(data_point))

    datasets_idx = np.zeros([2,NUM_COMPLEXITIES])
    for idx, opt in enumerate(opt_data):
        if not opt.num_images_training == 1e5:
            continue
        if opt.complexity>4:
            continue
        if opt.complexity_strict:
            datasets_idx[0,opt.complexity] = idx
        else:
            datasets_idx[1,opt.complexity] = idx
```

```python
#ORIGINAL
# im_sizes = [10,12,14,18,24]
im_sizes = [8,10,12,14,18]
dataset_test_labels = ['Polar','Spiral','Digs']
dataset_train_idx = [49, 50, 53]

result = np.zeros((5,3))
ids = np.zeros((5, 3))

for res in results_data_generalization[-1]:
    if not 'test_accuracy' in res['results']:
        print(res)
        continue
    idx = np.where(np.asarray(im_sizes) == res['im_size'] )[0][0]
    for idx_test, dataset in enumerate(dataset_train_idx):
        if result[idx, idx_test] < res['results']['test_accuracy'][dataset]:
            result[idx, idx_test] = res['results']['test_accuracy'][dataset]
            ids[idx, idx_test] = res['id']
```

```python
# im_sizes = [10,12,14,18,24]
im_sizes = [8,10,12,14,18]
iterations = [3, 5, 10, 20, 30]
im_size = 18
# im_sizes = [24, 42]
dataset_test_labels = ['Polar','Spiral','Digs']
dataset_train_idx = [49, 50, 53]
best_model_parameters = np.zeros((5, 3))

result = np.zeros((5,3))
for res in results_data_generalization[-1]:
    if not 'test_accuracy' in res['results']:
        print(res)
        continue
    idx = np.where(np.asarray(iterations) == res['iterations'] )[0][0]
    for idx_test, dataset in enumerate(dataset_train_idx):
        if res['im_size'] == im_size and result[idx, idx_test] <= res['results']['test_accuracy'][dataset]:
            result[idx, idx_test] = res['results']['test_accuracy'][dataset]
            best_model_parameters[idx, idx_test] = res['iterations']

```
